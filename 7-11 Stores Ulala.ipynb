```python
import json

import requests
from bs4 import BeautifulSoup # Import Beautiful Soup 4


# 7-Eleven Website URL we are going to call
url = "http://emap.pcsc.com.tw/EMapSDK.aspx"

# Dictionary the future JSON file will be from
jsonlist = []


# Iterating over the dataframe previously created
for index, row in taiwan.iterrows(): 
    # adding parameters from the dataframe to the payload
    payload = "commandid=SearchStore&city=" + row['City'].encode('utf-8') + "&town=" + row['TownName'].encode('utf-8')
    # headers required for the calls
    headers = {
        'cache-control': "no-cache",
        'postman-token': "f946b8c2-c86f-5906-e893-decf248936c8",
        'content-type': "application/x-www-form-urlencoded"
        }

    response = requests.request("POST", url, data=payload, headers=headers)
#     Making the call
    soup = BeautifulSoup(response.text, 'xml')
    
    # creating dictionary object for each store in specific town, to append to jsonlist
    for address, id, name, y, x, phone in zip(soup.find_all('Address'), soup.find_all('POIID'), soup.find_all('POIName'), soup.find_all('Y'), soup.find_all('X'), soup.find_all('Telno')):
        shops = {}
        # "sevenelevenshop" is the name of the Django model
        fields = {"model": "addresses.sevenelevenshop", 'fields': shops}
        shops['store_name'] = name.get_text()
        shops['store_id'] = id.get_text().replace(" ", "")
        shops['address'] = address.get_text()
        shops['town'] = row['TownName']
        shops['city'] = row['City']
        shops['y'] = str(float(y.get_text()) * 0.000001)
        shops['x'] = str(float(x.get_text()) * 0.000001)
        shops['phone'] = phone.get_text().replace(" ", "")
        jsonlist.append(fields)
                

# dumping jsonlist to a JSON file that can be used to create Django objects
# print >>f, json.dumps(jsonlist, ensure_ascii=False, indent=1)
```
